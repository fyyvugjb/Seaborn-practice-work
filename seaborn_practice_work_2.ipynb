```python
#4
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = {'pulse': [70, 75, 80, 85, 90, 95, 100, 72, 78, 84, 90, 96, 102, 68, 74, 80, 86, 92, 98],
        'diet': ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C']}
exercise_df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
for diet in exercise_df['diet'].unique():
    sns.histplot(exercise_df[exercise_df['diet'] == diet]['pulse'], label=diet, kde=False, element="step")
for diet in exercise_df['diet'].unique():
    sns.kdeplot(exercise_df[exercise_df['diet'] == diet]['pulse'], label=diet, linestyle='--')
plt.xlabel('Pulse Rate')
plt.ylabel('Frequency/Density')
plt.title('Distribution of Pulse Rates by Diet (Stacked Histogram and KDE)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#8
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
import seaborn as sns
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]
median_sepal_length = iris_df.groupby('species')['sepal length (cm)'].median()
plt.figure(figsize=(8, 6))
median_sepal_length.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Species')
plt.ylabel('Median Sepal Length (cm)')
plt.title('Median Sepal Length by Species')
plt.xticks(rotation=0)
plt.show()
```

```python
#12
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = {
    'Subject': ['Python', 'Python', 'Python', 'English', 'English', 'English', 'Robotics', 'Robotics', 'Robotics'],
    'Batch': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Performance': [85, 78, 92, 70, 82, 88, 90, 75, 85],
    'StdDev': [5, 3, 6, 4, 2, 5, 7, 3, 4]
}
df = pd.DataFrame(data)
subject_order = ['Python', 'English', 'Robotics']
colors = {'A': 'red', 'B': 'green', 'C': 'black'}
bar_width = 0.25
x = np.arange(len(subject_order))
fig, ax = plt.subplots(figsize=(10, 6))
for i, batch in enumerate(['A', 'B', 'C']):
  batch_data = df[df['Batch'] == batch]
  performance = batch_data.groupby('Subject').Performance.mean()
  std_dev = batch_data.groupby('Subject').StdDev.mean()
  positions = x + i * bar_width
  ax.bar(positions, performance[subject_order], bar_width, label=batch, yerr=std_dev[subject_order], color=colors[batch], capsize=5)
ax.set_xticks(x + bar_width)
ax.set_xticklabels(subject_order)
ax.set_xlabel('Subject')
ax.set_ylabel('Performance')
ax.set_title('Performance of Batches across Subjects')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
#14
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
diamonds = sns.load_dataset('diamonds')
```

```python
# Price vs. Carat with color
plt.figure(figsize=(8, 6))
sns.scatterplot(x='carat', y='price', data=diamonds, hue='color')
plt.title('Diamond Price vs. Carat (Colored by Diamond Color)')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
# Boxenplot of price distribution across clarity
plt.figure(figsize=(10,6))
sns.boxenplot(x='clarity', y='price', data=diamonds)
plt.title('Diamond Price Distribution by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Price')
plt.show()
```

```python
# Visualize the relationship between price and cut
plt.figure(figsize=(10,6))
sns.violinplot(x='cut', y='price', data=diamonds)
plt.title('Diamond Price Distribution by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```
